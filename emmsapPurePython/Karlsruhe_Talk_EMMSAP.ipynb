```python
%load_ext music21.ipython21
from __future__ import unicode_literals
from __future__ import print_function
from __future__ import division

from music21 import *
from emmsap import files, mysqlEM, indexRatios
```

```python
em = mysqlEM.EMMSAPMysql()
p = em.pieceByFilename('E15cM-VII_HS-4_Patrem.xml')
p.stream().measures(1, 6).show()
```

```python
segs = p.segments()
segs[0:2]
```

```python
s0 = segs[0]
s0.segmentData
```

```python
p2 = em.pieceById(1959)
p2.filename
```

```python
p2.stream().measures(1, 10).show()
```

```python
ratios = []
highParts = None
highRatio = 5000
for s1 in segs:
    for s2 in p2.segments():
        r = indexRatios.ratiosFromSegments(s1, s2)
        if r >= highRatio:
            highRatio = r
            highPart = p.partFromSegmentPair(s1, s2)
        ratios.append(r)
```

```python
max(ratios)
```
