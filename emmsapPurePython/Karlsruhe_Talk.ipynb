```python
c = corpus.parse('bwv66.6')
c
```

```python
c.show()
```

```python
cc = c.chordify()
cc.show()
```

```python
cc.transpose('P4')
cc.show()
```

```python
for ch in cc.flat.getElementsByClass('Chord'):
    ch.closedPosition(inPlace=True, forceOctave=4)
```

```python
cc.show()
```

```python
for ch in cc.flat.getElementsByClass('Chord'):
    ch.lyric = ch.commonName
```

```python
cc.show()
```

```python
chords = cc.flat.getElementsByClass('Chord')
for ch in chords:
    if ch.commonName not in ('major triad', 'minor triad'):
        ch.color = 'red'
cc.show()
```

```python
for ch in chords:
    if ch.commonName not in ('major triad', 'minor triad') and ch.beat != int(ch.beat):
        ch.color = 'black'
cc.show()
```
