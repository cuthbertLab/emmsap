```python
c = corpus.parse('bwv66.6')
```

```python
c.show()
```

```python
cc = c.chordify()
```

```python
cc.show()
```

```python
chords = cc.flat.getElementsByClass('Chord')
for ch in chords:
    ch.closedPosition(inPlace=True, forceOctave=4)
```

```python
cc.show()
```

```python
for ch in chords:
    if ch.isConsonant() is False and ch.beat not in [1,2,3,4]:
        ch.color = 'yellow'
```

```python
cc.show()
```
