## Extracting signatures

```python
laudamus = re.compile('laudamus te', re.IGNORECASE)
gratias = re.compile('gratias agimus', re.IGNORECASE)
quiTollis = re.compile('qui tollis', re.IGNORECASE)
quiSedes = re.compile('qui sedes', re.IGNORECASE)
cumSancto = re.compile('cum sancto', re.IGNORECASE)
```

```python
allProportions = []

for i, fp in enumerate(gloriaList):
    if i > 20:
        break
    gloriaScore = converter.parse(fp)
    firstPart = gloriaScore.parts[0]
    partFlat = firstPart.flat
    searcher = search.lyrics.LyricSearcher(firstPart)
    fileNameShort = fp.split('/')[-1][0:-4]
    offsets = []
    try:  # IndexErrors Galore!
        for rePattern in (laudamus, gratias, quiTollis, quiSedes, cumSancto):
            matches = searcher.search(rePattern)
            if matches:
                firstNote = matches[0].els[0]
            else:
                textExpressions = partFlat.getElementsByClass('TextExpression')
                for te in textExpressions:
                    if rePattern.match(te.content):
                        firstNote = te
                        print("YEP!")
            startOfSection = firstNote.getOffsetBySite(partFlat)
            offsets.append(startOfSection)
    except NameError: # raised if firstNote not found...
        continue
    denominator = offsets[1] - offsets[0]
    proportions = []
    for i in range(1, len(offsets) - 1):
        ratio = (offsets[i + 1] - offsets[i]) / denominator
        proportions.append(round(ratio, 3))
    proportions.append(fileNameShort)
    allProportions.append(proportions)
    print(proportions)
```

```python
allProportions[0:2]
```

```python
a0 = allProportions[0]
a1 = allProportions[1]
errorAmt = abs(a0[0] - a1[0]) + abs(a0[1] - a1[1]) + abs(a0[0] - a1[0])
errorAmt
```

```python
closestMatches = []

for a0 in allProportions:
    minError = 999
    minPiece = None
    for a1 in allProportions:
        if a0 is a1:
            continue
        errorAmt = abs(a0[0] - a1[0]) + abs(a0[1] - a1[1]) + abs(a0[0] - a1[0])
        if errorAmt < minError:
            minPiece = a1
            minError = errorAmt
    closestMatches.append([a0[-1], minPiece[-1], round(minError, 4)])
```

```python
sorted(closestMatches, key=lambda x: x[2])
```
