```python
c = corpus.parse('bwv66.6')
c.show()
```

```python
cc = c.chordify()
cc.show()
```

```python
for ch in cc.flat.getElementsByClass('Chord'):
    ch.closedPosition(inPlace=True, forceOctave=4)
cc.show()
```

```python
chords = cc.flat.getElementsByClass('Chord')
```

```python
for ch in chords:
    ch.lyric = ch.commonName

cc.show()
```

```python
for ch in chords:
    if ch.isConsonant() is False:
        ch.color = 'yellow'
```

```python
cc.show()
```

```python
for ch in chords:
    if ch.beat != int(ch.beat):
        ch.color = 'blue'
cc.show()
```

```python
from emmsap import files, mysqlEM, indexRatios
```

```python
em = mysqlEM.EMMSAPMysql()
p = em.pieceByFilename('E15cM-VII_HS-4_Patrem.xml')
```

```python
p.stream().measures(1,6).show()
```

```python
segs = p.segments()
segs
```

```python
s0 = segs[0]
s0.segmentData
```

```python
p.stream().parts[0].measures(1,4).show()
```

```python
p2 = em.pieceById(1959)
p2.filename
```

```python
p2.stream().measures(1,10).show()
```

```python
ratios = []
highPart = None
for s1 in segs:
    for s2 in p2.segments():
        r = indexRatios.ratiosFromSegments(s1, s2)
        if r >= 4700:
                highPart = p.partFromSegmentPair(s1, s2)
        ratios.append(r)
```

```python
max(ratios)
```

```python
highPart
```

```python
highPart.show()
```

```python
p = em.pieceByFilename('PMFC_12_9-Gloria_Clementiae_Pax.xml')
segs = p.segments()
```

```python
ratios = []
highPart = None
highRatio = 5000
for s1 in segs:
    for s2 in p2.segments():
        r = indexRatios.ratiosFromSegments(s1, s2)
        if r >= highRatio:
            highRatio = r
            highPart = p.partFromSegmentPair(s1, s2)
        ratios.append(r)
```

```python
max(ratios)
```

```python
highPart.show()
```
